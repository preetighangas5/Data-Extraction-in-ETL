**DATA EXTRAXTION IN ETL**

**Question 1:** Describe different types of data sources used in ETL using suitable examples ?

**Answer:** Data Sources in ETL(Extract, Transform, Load) are broadly classified into four categories:


*  Structured Data Sources:- These have a fixed schema and are highly organized.Examples- Relational Database (MY SQL, PostgreSQL, Oracle, SQL Server), data warehouse (Snowflake, Redshift).

*   Semi-Structured Data Sources:- These have some stuructures(tages or keys) but are flexible. Examples- JSON files, XML files, CSV/TSV files, Parquet/ORC FILES, NoSQL Databasse(Mongo DB, Cassandara).

*  UnstructureD Data Sources :- These have no predfined schema.
 Examples -Text documents, PDFs, E-mails, log files, images, videos, social media posts.

*  Application and Real-Time Sources :- These are accesses via interfaces rather than direct filesdatabases.Examples- REST/SOAP APIs(e.g., Sales force, Google Analytics), streaming platforms(kafka, kinesis), cloud storage (AWS S3, Azure Blop, Google Cloud Storage), Saas Applications(HubSpot, Shopify).


**Question 2:** What is Data Extraction ? Explain its role in the ETL pipeline.

**Answer:** Data Extraction is the first stage of ETL pipeline.It is the process of retrieving raw data from one or more heterogenous sourcce system and bringing it into ETL environment.

 Role in the ETL Pipeline:


*   Act as the entry point- without proper extraction, no transform or loading can occur.
*   Ensures data is collected efficiently, securely, and with minimal impact on source systems.


*   Handle Different formats, protocols and access methods.
*   Dtermine the overall performance and reliability of the entire ETL process(incremental vs full load, batch vs real-time, etc.)





**Question 3:** Explain the difference between CSV and Excel in terms of extraction and ETL usage.

**Answer:** CSV and Excel are two common file formats used in data extraction, but they are very different in terms of how they work and how they are used in ETL processes.
CSV is a simple plain-text file where data is separated by commas. It is very lightweight, fast to read, and has no practical limit on the number of rows. Because of this, it uses very little memory and can be processed quickly even for huge files. In ETL, CSV is the preferred choice for large-scale batch jobs because it is easy to stream and handle with basic tools.
Excel (.xlsx or .xls), on the other hand, is a more complex file format. It can contain multiple sheets, formatting, formulas, charts, and styles. However, it is heavier, slower to read, and has a limit of about 1 million rows per sheet. Reading Excel files usually requires special libraries and often loads the entire file into memory, which makes it slower for big data.
In short, use CSV when you need speed and simplicity in ETL. Use Excel only when the source data comes as a formatted business report with multiple sheets.

**Question 4:** Explain the steps involved in extracting data from a relational database.

**Answer:** The Standard steps are:


1.   Establish Connection - Use JDBC/ODBC driver or native connector (e.g., psycopg2 for PostgreSQL, mysql-connector for MySQL)

2.   Authentication & Authorization - Provide credentials and ensure the user has SELECT privileges.
3.   Define Extraction Logic
      - Full load: SELECT * FROM table
      - Incremental load: Use WHERE last_updated > ? or CDC (Change Data Capture)
4.  Execute Query- Run the SQL Statment.
5.  Fetch Data- Use batch/cursor fetching (e.g., fetchmany(10000)) to avoid loading everything into memory at once.
6.  Handle Errors & Logging- Implement retry logic, timeouts, and detailed logging.
7.  Close Connection - Always release resources.

Modern ETL tools (Airflow, Talend, Informatica, dbt) automate most of these steps.






**Question 5:** Explain three common challenges faced during Data Extraction.

**Answer:**
1.  Performance Impact on Source Systems
Running heavy SELECT queries on production databases can slowdownbusiness applications.
Solution: Use read replicas, schedule extractions during off-peak hours, or implement Change Data Capture (CDC).

2.  Data Volume & Scalability
Extracting terabytes of data can cause memory exhaustion or extremely long run times.
Solution: Process in chunks, use distributed frameworks (Spark), or incremental extraction.

3.  Data Quality & Consistency Issues
Sources often contain nulls, duplicates, encoding problems, schema changes, or inconsistent formats.
Solution: Add validation at extraction time, use schema-on-read techniques, and implement data profiling.

**Question 6:** What are APIs? Explain how APIs help in real-time data extraction.

**Answer:** APIs (Application Programming Interfaces) are sets of rules and protocols that allow different software applications to communicate with each other.
How APIs enable real-time data extraction:


*   REST/SOAP APIs- Pull-based extraction (ETL tool calls the API at regular intervals).
*   Webhooks- Push-based (source system sends data to ETL endpoint the moment an event occurs).


*   Streaming APIs (e.g., Kafka, WebSockets, Server-Sent Events) → Continuous real-time data flow.
*   Authentication (OAuth, API keys, JWT) ensures secure access.

Advantage over traditional methods: Instead of waiting for nightly batch files, data can be extracted within seconds or milliseconds, enabling near real-time analytics and operational dashboards.





**Question 7:** Why are Databases preferred for enterprise-level data extraction ?

**Answer:** Databases are preferred because they offer:


*   Structure & Integrity- Enforced schema, data types, constraints, and ACID properties.
*   Scalability - Handle millions of rows efficiently with indexing, partitioning, and sharding.


*   Security - Role-based access, encryption at rest/transit, audit logs.
*   Concurrency Control - Multiple users/queries can run simultaneously without data corruption.


*   Performance Optimization - Query optimization engines, materialized views, read replicas.
*   Change Tracking - Built-in CDC, transaction logs, and triggers for incremental extraction.


*   Tool Integration - Almost every ETL/ELT tool has native connectors for major databases.

Flat files or spreadsheets cannot match these capabilities at enterprise scale.







**Question 8:** What steps should an ETL developer take when extracting data from large CSV files(1GB+) ?

**Answer:** Follow these best practices:

 1. Never load the entire file into memory - Use streaming/chunking.
 2. Choose the right library:
    -  Python: pandas with chunksize, Dask, Polars (very fast), or pure csv module with generators.
    -  Spark / Databricks: spark.read.csv() with distributed processing.
 3. Validate file properties first - Check encoding (UTF-8), delimiter, quote character, and header.
 4. Process in chunks - Read 10,000-100,000 rows at a time, transform, and write to staging.
 5. Use parallel processing - Split the file (if possible) or use multi-threading / Dask/Spark.
 6. Monitor resources - Track memory and CPU usage; implement checkpointing.
 7. Handle errors gracefully - Log bad rows and continue processing.
 8. Compress if needed - Work with gzipped CSVs directly (most tools support it).
